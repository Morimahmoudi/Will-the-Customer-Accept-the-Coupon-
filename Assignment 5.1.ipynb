```python
#importing the required libraries for the analysis

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# import the data from the CSV file

DataFrame = r"C:\Users\morte\Downloads\assignment5_1_starter\data\coupons.csv"
df = pd.read_csv(DataFrame)
df
```

```python
#cleaning the data in column "Y" to ensure all rows contain "0" or "1"

df_clean = df[df["Y"].isin([0, 1])]
df_clean
```

```python
# identifying various tyoes of resturant and bars in "coupon" column and defining the sum of numbers in "Y" column which shows the numbers of individuals who accepted the coupan or differed in to a later time 

df_grouped = df.groupby("coupon")[["Y"]].sum()
df_grouped
```

```python
# Barplot comparing the sum of accepted coupons in each coupon category 

df_grouped.plot(kind = "bar") 
plt.title("Distribution of the sum of the accepted coupans by their type")
plt.xlabel("Restaurant type")
plt.ylabel("Numbers of accepted coupons")
plt.savefig("C:/Users/morte/Dropbox/Files/Harvard Files/Desktop-Harvard-May2016/Professorship/ML course Berkeley/Figures/coupon.png", dpi=300)
plt.show()
```

```python
# Identifying the passengers (e.g., whether they were alone or with someone else) group and the sume of acceted and rejected coupons in each category 

df_passanger = df.groupby("passanger")[["Y"]].value_counts()
df_passanger
```

```python
# Identifying the numbers of accepted and rejected coupons in each coupon type and passenger group 

df_copass = df.groupby(["coupon" , "passanger"])[["Y"]].value_counts()
df_copass
```

```python
# Bar plots comparing the numbers of accepted and rejected coupons in each coupon type and passenger group 

df_copass = df.pivot_table(index=["coupon", "passanger"], columns="Y", aggfunc="size", fill_value=0)
df_copass
df_copass.plot(kind="bar")
plt.title("Distribution of accepted or rejected coupons by their type and passenger group")
plt.xlabel("Restaurant type and and passenger group")
plt.ylabel("Numbers of accepted and rejected coupons")
plt.legend(title="Accepted (1); Rejected (0) ")
plt.savefig("C:/Users/morte/Dropbox/Files/Harvard Files/Desktop-Harvard-May2016/Professorship/ML course Berkeley/Figures/passenger accepted rejected.png", dpi=300)
plt.show()
```

```python
# Table summarizing the numbers of accepted and rejected coupons together with the percentage of coupon acceptance in each coupon type and passenger group 

df_copass["% of getting coupon"] = (df_copass[1] / (df_copass[1] + df_copass[0])) * 100
df_copass
```

```python
# Bar plots comparing percentage of accepted coupons in each coupon type and passenger group 

plt.figure(figsize=(12,6))
sns.barplot(x="coupon", y="% of getting coupon", hue="passanger", data=df_copass)
plt.title("Distribution of accepted coupons by their type and passenger category")
plt.xlabel("Restaurant type")
plt.ylabel("Percentage of accepted coupons")
plt.legend(title="Passenger category")
plt.savefig("C:/Users/morte/Dropbox/Files/Harvard Files/Desktop-Harvard-May2016/Professorship/ML course Berkeley/Figures/passenger.png", dpi=300)
plt.show()
```
